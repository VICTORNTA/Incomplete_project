```python
import tensorflow as tf
from tensorflow import image
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dropout,Dense
from tensorflow.keras.models import  Sequential
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow.keras.applications import EfficientNetB6
```

```python
batch_size = 64
```

```python
train_dir = r"/content/drive/MyDrive/army_worm_project/armyworm_project/train_images/resized_images"
```

```python
train_data_gen =ImageDataGenerator(
    rescale=1./255, 
    horizontal_flip=True,
    rotation_range = 120,
    validation_split = 0.3
)
```

```python
train_image_gen = train_data_gen.flow_from_directory(
    batch_size = batch_size,
    directory = train_dir,
    target_size = (IMG_SIZE,IMG_SIZE),
    class_mode = 'categorical',
    subset = "training"

    
)
```

```python
validation_generator = train_data_gen.flow_from_directory(
    train_dir, 
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation') # set as validation data
```

```python
model = EfficientNetB6(include_top=False,weights='imagenet',input_shape=(IMG_SIZE,IMG_SIZE,3))
```

```python
sample_train_image,c = next(validation_generator)
```

```python
import matplotlib.pyplot as plt
```

```python
fig,axes = plt.subplots(5,5,figsize=(20,20))
axes = axes.flatten()
for _,img in zip(axes,sample_train_image[25:56]):
  _.imshow(img) 
  _.axis("off")
```

```python
model.trainable = False
```

```python
model.output
```

```python
from tensorflow.keras import layers
x = layers.GlobalAveragePooling2D(name="avg_pool")(model.output)
x = layers.BatchNormalization()(x)
top_dropout_rate = 0.2
x = layers.Dropout(top_dropout_rate, name="top_dropout")(x)
outputs = layers.Dense(1, activation="softmax", name="pred")(x)
```

```python
model.input
```

```python
model = tf.keras.Model( model.inputs, outputs, name="Efficientnetb6")
model.output
```

```python
model.compile(
    optimizer ='adam',
    loss = "binary_crossentropy",
    metrics = ['accuracy'],

)
```

```python
model.summary()
```

```python
model.fit(
    train_image_gen,
    steps_per_epoch=1134//64,
    epochs = 10,
    validation_data = validation_generator
    )
```
